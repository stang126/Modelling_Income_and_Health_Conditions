## Background & Prior Work

$\;\;\;\;\;\;$ Poverty has long been recognized as a contributor to disease and death. Income has been strongly correlated with life expectancy for decades. In the 1970s, a 60 year old man in the top half of the income distribution was expected to live 1.2 years longer than a man in the bottom half. By the end of the century, he could expect to live 5.8 years longer <sup>[1]</sup>. Income influences health in various ways which can be loosely categorized as clinical, behavioral, and environmental, the latter two being closely related.

$\;\;\;\;\;\;$ Regarding the clinical factors, low-income individuals face greater barriers to accessing medical care than higher-income individuals. Low-income workers are more likely to be employed by organizations that do not offer health benefits: Less than one-third of low-income workers obtain health insurance through their employer, compared to nearly 60 percent of higher-income workers <sup>[1]</sup>. People without healthcare are more likely to forgo medical care due to cost concerns.

$\;\;\;\;\;\;$ As for the behavioral and environmental factors, lower-income Americans have higher rates of behavioral risk factors such as smoking, obesity, substance use, and low levels of physical activity. Targeted marketing of tobacco companies in low-income neighborhoods and limited access to counseling services for low-income people have contributed to greater smoking rates in lower income households than higher income households <sup>[1]</sup>. Environmental pollutants and poorer living standards in low-income neighborhoods may also contribute to this disparity in health. Structural challenges within these neighborhoods have contributed to increased obesity. A higher density of fast-food restaurants, limited access to fresh foods, and fewer open-space environments that are conducive to physical activity within these neighborhoods severely limit one's ability to maintain a healthy lifestyle.

$\;\;\;\;\;\;$ In the past few years, the rates of obesity within the United States have increased while the wealth gap has also increased. While no correlation between obesity and economic class may have existed previously in the United States, changes in lifestyle over the past decade could have caused this trend to change. Due to the nature of mass production of processed foods, unhealthy food is less expensive and available than their healthier counterparts. As a result, these foods are more accessible for people of lower income households. Fructose, for example, is a commonly used sweetener in beverages that decreases insulin sensitivity in obese people <sup>[1]</sup>. While overall sugar consumption gradually rose in the 20th century, from 12 percent of U.S. food energy in 1909 to 19 percent by the year 2000, the use of high fructose corn syrup (HCFS) in the U.S. increased from virtually zero per capita in 1970 to over 60 pounds per capita annually in the U.S. in 2000, about half of total sugar consumption <sup>[1]</sup>. HFCS is now the main sweetener in sweetened beverages, which as of 2016, have constituted over 7 percent of household food expenditures and over 9 percent of expenditures for low-income households in the SNAP program <sup>[2]</sup>.  
    
$\;\;\;\;\;\;$ Across different states, there is a general geographic heterogeneity of obesity rates in the US <sup>[2]</sup>. In recent years, an inverse correlation has been identified between obesity and household income. In 2015, in states with a median household income below $\$$45,000 per year, such as Alabama, Mississippi, and West Virginia, over 35 pecent of the population was obese; meanwhile in states with a median household earning of more than $\$$65,000 per year, obesity rates were lower than 25 percent. While financial constraints may help to explain this trend, the Social Learning hypothesis provides an alternate explanation. Local communities can have their own cultural and social norms which may therefore underlie local differences in obesity and diabetes rates.
    
$\;\;\;\;\;\;$ In addition to geography as a confounding variable in the correlation between health factors and income level, race and ethnicity may influence this relationship as well. In a study from the CDC in 2010, obesity in men was found to be similar across income levels with a slightly higher rate among higher income levels; however, these trends differ between race and ethnicity groups. For non-Hispanic black and Mexican American men, the obesity rate tended to decrease as income levels decreased, but there was no significant difference between obesity rates across income levels in non-Hispanic white men. Additionally, it was found that women in more affulent households had lower obesity rates than their lower income counterparts <sup>[3]</sup>. 

$\;\;\;\;\;\;$ According to the CDC, obese individuals are at an increased risk for developing more serious diseases and health conditions than normal weight individuals. Some of these conditions include: Type 2 diabetes, high cholesterol, heart disease, stroke, and sleep apnea <sup>[4]</sup>. Like obesity, these conditions can likely be attributed to poorer clinical and environmental conditions in low-income communities.

$\;\;\;\;\;\;$ While these findings were from highly trustworthy sources, such as Nature and the CDC, another study from the NIH that examine the trend between obesity and income found that publication bias can affect these trends. While the meta-analyses of the data suggested a significant correlation bewtween obesity and income level, adjusting for publication bias resulted in this correlation vanishing <sup>[5]</sup>. Due to these conflicting results, the correlation between obesity and economic status may need to be further studied. However, regardless of the relationship between obesity and income exclusively, income and general physical health appear to have some trend.

# Hypothesis


There is a connection between income and health that can be tied to various clinical and environmental factors such as varying access to healthcare, to healthy foods, and to environments conducive to healthy lifestyles. With lower income individuals possessing less access to these benefits, it is more difficult for them to lead healthy lifestyles. As a result, a trend has developed wherein lower income individuals tend to have a greater prevalance of health issues such as obesity and heart disease. Due to the heterogeneity of income across the United States, we believe geographic location can be a confounding variable in this trend as well. Because of the apparent trend between health conditions and income, we hypothesize that we will be able to predict one's income, given one's various health conditions and geographic location.

# Setup

```python
# import matplotlib
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.style as style

# set plotting size parameter
plt.rcParams['figure.figsize'] = (17, 7)

# Statmodels & patsy
import patsy
import statsmodels.api as sm

import pandas as pd
import numpy as np
import seaborn as sns
sns.set()
sns.set_context('talk')

import warnings
warnings.filterwarnings('ignore')

# scikit-learn imports
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor

# load data
CDC_health_df = pd.read_csv('https://chronicdata.cdc.gov/api/views/kee5-23sr/rows.csv')
IRS_income_df = pd.read_csv('https://www.irs.gov/pub/irs-soi/19zpallnoagi.csv')
zipcodes_df = pd.read_csv('uszips.csv')
```

# Data Cleaning

- The CDC dataset (`CDC_health_df`) will be truncated to show the zipcode, population in 2010, and a model-based estimate for the crude prevalence of obesity, diabetes, high cholesterol, sleep deprivation, stroke, and kidney disease in adults; each will be renamed for clarity. Additionally, we will drop any rows with missing information.

```python
#Load CDC Obesity Data by Zipcode
CDC_health_df = CDC_health_df[
    [
        'ZCTA5','TotalPopulation',
        'OBESITY_CrudePrev',
        'DIABETES_CrudePrev',
        'HIGHCHOL_CrudePrev',
        'SLEEP_CrudePrev',
        'STROKE_CrudePrev',
        'KIDNEY_CrudePrev']
]
CDC_health_df.columns = [
    'Zipcode',
    'Population',
    'Obesity',
    'Diabetes',
    'HighChol',
    'Sleep',
    'Stroke',
    'Kidney'
]
CDC_health_df=CDC_health_df.dropna()
CDC_health_df.head()
```

- The IRS dataset (`IRS_income_df`) will be truncated to show the state, zipcode, total number of returns filed, and the total income amount; each of these column names were renamed for clarity. Additionally, we will drop any rows with missing information.

```python
#Load Income data and # of tax returns by State and Zipcode
IRS_income_df = IRS_income_df[['STATE','ZIPCODE','N1','A02650']]
IRS_income_df.columns = ['State','Zipcode','Returns','Income']
IRS_income_df = IRS_income_df.dropna()
IRS_income_df.head()
```

- The Zipcode dataset (`zipcodes_df`) will be truncated to show the zipcode, city, and state abbreviation; each of these column names were renamed for clarity. Additionally, we will drop any rows with missing information.

```python
#Load Zipcode data
zipcodes_df = zipcodes_df[['zip','city','state_id']]
zipcodes_df.columns = ['Zipcode','City','State']
zipcodes_df = zipcodes_df.dropna()
zipcodes_df.head()
```

- The IRS dataset and the zipcode data set were combined to map a zipcode to its city (`city_income_df`); this allows us to relate the income of different parts of a city to the entire city. Additionally, as the IRS income data is reported as the total income that zipcode makes in the thousands, it was divided by the number of returns and multiplied by 1000 to understand the average income of a city's population.

```python
#Merge Zipcodes and Cities to get income by Zipcode
city_income_df = pd.merge(IRS_income_df, zipcodes_df, on = ['Zipcode','State'])
city_income_df.dropna(inplace = True)
city_income_df['Income'] = (city_income_df['Income'] / city_income_df['Returns']) * 1000
city_income_df.drop('Returns', axis = 1, inplace = True)
city_income_df.head()
```

- The city-income dataset (`city_income_df`) was then combined with the CDC dataset (`CDC_health_df`) to allow us to relate a city's income to its different health prevalences. Finally, the average incomes (based on the zipcode) for a city was then averaged to determine a city's average income. Additionally, the income data was normalised such that the income values range from 0 to 100.

```python
#Merge Obesity and Income Data
income_obesity_df = pd.merge(city_income_df, CDC_health_df, on = ['Zipcode'])
income_obesity_df = income_obesity_df[
    [
        'City',
        'State',
        'Zipcode',
        'Income',
        'Population',
        'Obesity',
        'Diabetes',
        'HighChol',
        'Sleep',
        'Stroke',
        'Kidney'
    ]
]
income_obesity_df.head()
```

```python
#Average Income by city
income_obesity_df = income_obesity_df.groupby(['State', 'City'], as_index = False).mean()
income_obesity_df['Income'] = (income_obesity_df['Income'] / max(income_obesity_df['Income'])) * 100
income_obesity_df.head()
```

Additionally, this model relies on the assumption that there is no multicollinearity within the data. However, as many of the health statistics are correlated to each other in some fashion, our data does have multicollinearity. 

In addition to examining the OLS regression $R^{2}$ value, we also plotted the predicted income values against the in sample true income values.

```python
# Comparing prediction and in sample true income
OLS_predictions = res_log.predict(predictors)
plt.scatter(x = OLS_predictions, y = outcome) # cant compare to the other 2
plt.ylabel('True Income')
plt.xlabel('Predicted Income')
plt.title('Multivariate Linear Regression True Vs Predicted Income')
```

Next, we used random forest regression to try to predict income based on state, population, and health statistics. In this, we tested using various number of trees in the forest as well as splitting the data in various ways.

To start, we did a one-hot encoding of the states; as states are considered categorical, we did a one-hot encoding to convert the states into a  numerical representation. 

```python
# One-hot encoding
ohe_pd = pd.concat([income_obesity_df, pd.get_dummies(income_obesity_df['State'])], axis = 1)
ohe_pd.head()
```

We then split our data into the predictors (X Variable) and outcome (Y variable). Our predictor dataframe, `stats_x`, contains the one-hot encoding of the states, population and obesity, diabetes, high cholesterol, sleep deprivation, stroke, and kidney disease prevalence. Our outcome array, `stats_y`, contains all the income information.

We then allocated 80% of our dataset for training data and the remaining 20% for testing data. Our resulting training variables are `stats_train_X` and `stats_train_Y`. Our resulting testing variables are `stats_test_X` and `stats_test_Y`. We then used a random forest regressor to model our data. In order to maximise our accuracy, we used varying numbers of trees in the forest.

```python
# Splitting data
stats_x = ohe_pd.iloc[:,3:]
stats_y = np.array(ohe_pd[['Income']])
stats_train_X, stats_test_X, stats_train_Y, stats_test_Y = train_test_split(stats_x, stats_y, 
                                                                            test_size = 0.2, 
                                                                            random_state = 200)
# Finding the best number of trees to use
mse = []
num_trees = [50, 100, 500, 1000, 1500]
for i in num_trees:
    rf = RandomForestRegressor(n_estimators = i, random_state = 42)
    rf.fit(stats_train_X, stats_train_Y)
    predictions = rf.predict(stats_test_X)
    mse.append(np.mean((predictions - stats_test_Y)**2))
```

Based on the number of trees in the random forest regression that minimise the mean squared error (MSE), we used that value to retrain our regression model. Based on this, we determined what the $R^{2}$ value was. 

```python
#Rerunning RFR algorithm
rf = RandomForestRegressor(n_estimators = num_trees[np.argmin(mse)], random_state = 42)
rf.fit(stats_train_X, stats_train_Y)
RFR_predictions = rf.predict(stats_test_X)
RFR_rs = rf.score(stats_test_X, stats_test_Y, sample_weight = None)
print('Random Forest Regression Number of Trees:', num_trees[np.argmin(mse)])
print('Random Forest Regression Mean Squared Error: %.4f' % min(mse))
print('Random Forest Regression R\u00b2 Value: %.4f' % RFR_rs)
```

As seen from the random forest regression results above, our model has an $R^{2}$ value of 0.9962 and a mean squared error value of 7.4462. Such a high $R^{2}$ value indicates that our random forest regression model with 50 trees is a good fit to use to predict income. Additionally, our MSE value is fairly low. 

In addition to examining the random forest regression $R^{2}$ value and MSE value, we also plotted the predicted income values against the true income values. 

```python
# Comparing True and Predicted Income 
plt.scatter(x = RFR_predictions, y = stats_test_Y)
plt.ylabel('True Income')
plt.xlabel('Predicted Income')
plt.title('Random Forest Regression True Vs Predicted Income')
```

As shown in the plot above, it appears that the relationship between the true and predicted income is very linear, where most predicted income values tend to reflect its true income value. From this, it does make sense why the $R^{2}$ value is very high and the MSE value is very low. 

Finally, we used the k-nearest neighbour algorithm to try to predict income based on population and the health statistics. In this, we tested using diffrent k values as well as splitting the data in various ways. 

To start, we initialised two lists, one will contain a k value and the other will contain its mean squared error. We also created a list of random states which will be used the shuffle the data before splitting it. We once again allocated 80% of our dataset for training data and the remaining 20% for testing data for each random state. The resulting training variables are `stats_train_X` and `stats_train_Y`. Our resulting testing variables are `stats_test_X` and `stats_test_Y`. 

We will then loop through various k values for our k-nearest neighbour algorithm to determine which k value will result in the lowest mean squared error for that data. The lowest mean squared error and its k value will be appended to the two lists we first initialised. 

```python
# Initialise best_k and best_mse lists
best_k = []
best_mse = []
random_state = [10, 50, 100, 150, 200, 250, 500, 550, 700]

# Finding the best k value and random state to use
for i in range(len(random_state)):
    stats_train_X, stats_test_X, stats_train_Y, stats_test_Y = train_test_split(stats_x, stats_y, 
                                                                                test_size = 0.2, 
                                                                                random_state = random_state[i])
    mse = []
    for k in range(1,100):
        neigh = KNeighborsRegressor(n_neighbors = k)
        neigh.fit(stats_train_X, stats_train_Y)
        predicted_y = neigh.predict(stats_test_X)
        mse.append(np.mean((predicted_y - stats_test_Y) ** 2))
    best_k.append(np.argmin(mse))
    best_mse.append(min(mse))
```

```python
# Rerunning KNN algorithm
random_state = random_state[np.argmin(best_mse) + 1]
stats_train_X, stats_test_X, stats_train_Y, stats_test_Y = train_test_split(stats_x, stats_y, 
                                                                                test_size = 0.2, 
                                                                                random_state = random_state)
k = best_k[np.argmin(best_mse) + 1]
neigh = KNeighborsRegressor(n_neighbors = k)
model = neigh.fit(stats_train_X, stats_train_Y)
KNN_predictions = neigh.predict(stats_test_X)
KNN_rs = neigh.score(stats_test_X, stats_test_Y, sample_weight = None)
print('K-Nearest Neighbor Algorithm K value:', k)
print('K-Nearest Neighbor Algorithm Mean Squared Error: %.4f' % min(best_mse))
print('K-Nearest Neighbor Algorithm R\u00b2 Value: %.4f' % KNN_rs)
```

As seen from the K Nearest Neighbour model results above, our model has an $R^{2}$ value of 0.1394 and a mean squared error of 2.3579. While the MSE value is fairly low, which normally indicates there is little error in the predicted values, our $R^{2}$ value is very low. Such a low $R^{2}$ value indicates that our K Nearest Neighbour model is not a good fit to use to predict income from state, population, and health statistics. 

In addition to examining the K Nearest Neighbour model $R^{2}$ value and MSE value, we also plotted the predicted income values against the true income values.

```python
# Comparing True and Predicted Income
plt.scatter(x = KNN_predictions, y = stats_test_Y)
plt.ylabel('True Y values')
plt.xlabel('Predicted Y values')
plt.title('K Nearest Neighbour True Vs Predicted Income')
```

As shown in the plot above, it appears that the relationship between the true and predicted income for the K Nearest Neighbour model is vaguely linear; however, most predicted income values do not reflect its true income value very well. From this, it does make sense why the $R^{2}$ value for this model is very low but it does not explain why the MSE is very low.

# Ethics & Privacy

After a preliminary negative correlation was established between income and prevalence of health conditions, we posed our question in a way that seeks to investigate whether health conditions and geographical location were sufficient markers to predict income.

All of the information used in our data analysis is from publicly available datasets from reliable sources at a large scale. We aimed to minimize any bias and strip all information that could be personally identifiable or used to target groups based on external factors other than location (such as gender, racial background, etc) by restricting the variables used from this dataset to the minimum amount necessary to perform our calculations.

To evaluate our project, our metric of success is the $R^2$ value of our models which is a consistent and quantifiable value. To avoid overfitting our data, we randomly split our data into training and testing datasets.

As discussed in our background work and conclusion, we reduced the chances of spurious correlation by using factors with genuine significance that we supported with our reasoning.

Some ethical and privacy concerns we considered during the project include:
- Our data is incomplete and does not account for the portion of the population that fails to file tax returns. If there is a correlation between individuals not filing taxes and income levels, this could potentially shift the results if a significant percentage fails to file taxes.
- The CDC dataset used provided us with ZCTA (Zipcode Tabulation Areas) rather than zipcodes. In certain cases, such as zipcodes with only businesses, or largely unpopulated areas, ZCTA values may not align exactly with the zipcodes.
- Businesses or other entities could potentially use this data to prey on individuals of certain income groups for their benefit, such as food companies promoting products in areas with lower income since they have higher rates of diabetes.
- Taking the average income for an entire zipcode may lump together many different groups of high and low incomes into a single value that does not accurately represent any of the individuals in that area.
